```
public class Solution
{
    public int MinAreaRect(int[][] points)
    {
        var xDict1 = new SortedDictionary<int, SortedSet<int>>();
        var yDict1 = new SortedDictionary<int, SortedSet<int>>();

        foreach(var point in points)
        {
            if(xDict1.ContainsKey(point[0]))
                xDict1[point[0]].Add(point[1]);
            else xDict1[point[0]] = new SortedSet<int> { { point[1] } };
            if(yDict1.ContainsKey(point[1]))
                yDict1[point[1]].Add(point[0]);
            else yDict1[point[1]] = new SortedSet<int> { { point[0] } };
        }

        var xDict2 = new Dictionary<int, List<int>>(xDict1.Select(x => KeyValuePair.Create<int, List<int>>(x.Key, x.Value.ToList())));
        var yDict2 = new Dictionary<int, List<int>>(yDict1.Select(x => KeyValuePair.Create<int, List<int>>(x.Key, x.Value.ToList())));

        int ret = 0;

        foreach(var kv in xDict2)
        {
            if(kv.Value.Count < 2)
                continue;
            
            int x1 = kv.Key;
            for(int i = 0; i < kv.Value.Count - 1; i++)
            {
                int y1 = kv.Value.ElementAt(i);
                if(yDict2[y1].Count < 2)
                    continue;

                for(int j = i + 1; j < kv.Value.Count; j++)
                {
                    int y2 = kv.Value.ElementAt(j);
                    if(ret > 0 && Math.Abs(y2 - y1) >= ret)
                        continue;
                    if(yDict2[y2].Count < 2)
                        continue;

                    int k = 0, l = 0;
                    while(k < yDict2[y1].Count && l < yDict2[y2].Count)
                    {
                        int px1 = yDict2[y1].ElementAt(k);
                        if(px1 == x1)
                        {
                            k++;
                            continue;
                        }
                        int px2 = yDict2[y2].ElementAt(l);
                        if(px2 == x1)
                        {
                            l++;
                            continue;
                        }
                        if(px1 == px2)
                        {
                            int x2 = px1;
                            int area = Math.Abs(x2 - x1) * Math.Abs(y2 - y1);
                            if(ret == 0)
                                ret = area;
                            else ret = Math.Min(ret, area);

                            l++;
                            k++;
                        }
                        else if(px1 < px2)
                            k++;
                        else l++;
                    }
                }
            }
        }

        return ret;
    }
}
```
