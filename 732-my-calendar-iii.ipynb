[732. My Calendar III](https://leetcode.com/problems/my-calendar-iii/)

```csharp
public class MyCalendarThree
{
    private List<(int start, int end, int count)> _list = new();
    
    public int Book(int start, int end)
    {
        // Console.WriteLine($"Inserting {start}->{end}");

        if(_list.Count == 0)
        {
            _list.Add((start, end, 1));
            // Console.WriteLine(string.Join(", ", _list.Select(x => $"{x.start}->{x.end} ({x.count})")));
            return 1;
        }
        
        int left = 0;
        int right = _list.Count;
        int startIndex = 0;
        while(left != right)
        {
            startIndex = left + (right - left) / 2;
            if(_list[startIndex].end <= start)
                left = startIndex + 1;
            else if(start < _list[startIndex].start)
                right = startIndex;
            else break;
        }

        if(_list[startIndex].end <= start)
        {
            if(startIndex == _list.Count - 1)
            {
                _list.Add((start, end, 1));
                // Console.WriteLine(string.Join(", ", _list.Select(x => $"{x.start}->{x.end} ({x.count})")));
                return _list.Select(x => x.count).Max();
            }
            else startIndex++;
        }
        
        // Console.WriteLine($"startIndex: {startIndex}");

        var copy = new List<(int start, int end, int count)>();
        copy.AddRange(_list);

        for(int i = startIndex; i < copy.Count; i++)
        {
            var el = copy[i];

            // Console.WriteLine($"start: {start} end: {end}");
            // Console.WriteLine($"Element: {el.start}->{el.end} ({el.count}, {i} < {_list.Count})");

            if(start < el.start)
            {
                if(end <= el.start)
                {
                    copy.Insert(i, (start, end, 1));
                    break;
                }
                else if(end < el.end)
                {
                    copy.Insert(i, (start, el.start, 1));
                    copy.Insert(i + 1, (el.start, end, el.count + 1));
                    copy[i + 2] = (end, el.end, el.count);
                    break;
                }
                else if(end == el.end)
                {
                    copy.Insert(i, (start, el.start, 1));
                    copy[i + 1] = (el.start, el.end, el.count + 1);
                    break;
                }
                else
                {
                    copy.Insert(i, (start, el.start, 1));
                    copy[i + 1] = (el.start, el.end, el.count + 1);
                    start = el.end;
                }
            }
            else if(start == el.start)
            {
                if(end < el.end)
                {
                    copy.Insert(i, (start, end, el.count + 1));
                    copy[i + 1] = (end, el.end, el.count);
                    break;
                }
                else if(end == el.end)
                {
                    copy[i] = (el.start, el.end, el.count + 1);
                    break;
                }
                else
                {
                    copy[i] = (el.start, el.end, el.count + 1);
                    start = el.end;
                }
            }
            else if(start < el.end)
            {
                if(end < el.end)
                {
                    copy.Insert(i, (el.start, start, el.count));
                    copy.Insert(i + 1, (start, end, el.count + 1));
                    copy[i + 2] = (end, el.end, el.count);
                    break;
                }
                else if(end == el.end)
                {
                    copy.Insert(i, (el.start, start, el.count));
                    copy[i + 1] = (start, el.end, el.count + 1);
                    break;
                }
                else
                {
                    copy.Insert(i, (el.start, start, el.count));
                    copy[i + 1] = (start, el.end, el.count + 1);
                    start = el.end;
                }
            }
        }

        if(copy.Last().end <= start)
            copy.Add((start, end, 1));

        _list = copy;

        // Console.WriteLine(string.Join(", ", _list.Select(x => $"{x.start}->{x.end} ({x.count})")));
        
        return _list.Select(x => x.count).Max();
    }
}
```
