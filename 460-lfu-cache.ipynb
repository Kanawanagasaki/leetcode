[460. LFU Cache](https://leetcode.com/problems/lfu-cache/)

```csharp
public class Element
{
    public int Key;
    public int Value;
    public int UseCounter;
    public int OrderIndex;
}

public class LFUCache
{
    private int _capacity;
    private Dictionary<int, Element> _cache = new();
    private List<Element> _order = new();
    
    public LFUCache(int capacity)
    {
        _capacity = capacity;
    }
    
    public int Get(int key)
    {
        if(!_cache.ContainsKey(key))
            return -1;
        
        var el = _cache[key];
        
        el.UseCounter++;
        UpdateOrder(el);
        return el.Value;
    }
    
    public void Put(int key, int value)
    {
        if(_capacity <= 0)
            return;

        if(_cache.ContainsKey(key))
        {
            var el1 = _cache[key];
            el1.Value = value;
            el1.UseCounter++;
            UpdateOrder(el1);
            return;
        }
        
        if(_cache.Count >= _capacity)
        {
            _cache.Remove(_order.Last().Key);
            _order.RemoveAt(_order.Count - 1);
        }
        
        var el = new Element
        {
            Key = key,
            Value = value,
            UseCounter = 1,
            OrderIndex = _order.Count
        };
        _order.Add(el);
        _cache[key] = el;
        UpdateOrder(el);
    }
    
    private void UpdateOrder(Element el)
    {
        int elIndex = el.OrderIndex;
        int toIndex = 0;

        int left = 0;
        int right = el.OrderIndex;

        while(right - left >= 0)
        {
            int index = left + (right - left) / 2;
            if(_order[index].UseCounter > el.UseCounter && _order[index + 1].UseCounter <= el.UseCounter)
            {
                toIndex = index + 1;
                break;
            }
            if(_order[index].UseCounter > el.UseCounter)
                left = index + 1;
            else right = index - 1;
        }

        _order.RemoveAt(el.OrderIndex);
        _order.Insert(toIndex, el);
        
        for(int i = toIndex; i <= elIndex; i++)
            _order[i].OrderIndex = i;
    }
}

/**
 * Your LFUCache object will be instantiated and called as such:
 * LFUCache obj = new LFUCache(capacity);
 * int param_1 = obj.Get(key);
 * obj.Put(key,value);
 */

```
